```python
import os
import json
import pandas as pd

root_dir = "/home/jk/colon_dynamic/output/test_knn_lrspatial_cemb_drds_tv0.001"   # <-- change this

rows = []

for dirpath, dirnames, filenames in os.walk(root_dir):
    if "results.json" in filenames:
        results_path = os.path.join(dirpath, "results.json")
        
        with open(results_path, "r") as f:
            data = json.load(f)
        
        # your structure has a single key like "ours_30000"
        key = list(data.keys())[0]
        metrics = data[key]
        
        rows.append({
            "folder": os.path.basename(dirpath),
            "PSNR": metrics.get("PSNR"),
            "SSIM": metrics.get("SSIM"),
            "LPIPS": metrics.get("LPIPS"),
            "DEPTH_MSE": metrics.get("DEPTH_MSE"),
        })

df = pd.DataFrame(rows).set_index("folder").sort_index()
df.loc["MEAN"] = df.mean(numeric_only=True)


df
```
